```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import csv, re

url = "https://jc.uol.com.br/"
r = requests.get(url, timeout=30)
if r.status_code != 200:
    print(f"HTTP {r.status_code}")
    raise SystemExit

soup = BeautifulSoup(r.content, "html.parser")
p_data = re.compile(r"/\d{4}/\d{2}/\d{2}/")

vistos, items = set(), []
for a in soup.select("a[href]"):
    h = urljoin(url, a.get("href"))
    if not h.startswith(url):
        continue
    if not p_data.search(h):
        continue
    t = a.get_text(strip=True)
    if not t or h in vistos:
        continue
    m = p_data.search(h)
    data = m.group(0).strip("/").replace("/", "-") if m else ""
    vistos.add(h)
    items.append({"titulo": t, "link": h, "data": data})

for i, it in enumerate(items, 1):
    print(f'{i:02d}. {it["titulo"]}{"  ["+it["data"]+"]" if it["data"] else ""}\n    {it["link"]}')

with open("jc_manchetes.csv", "w", newline="", encoding="utf-8-sig") as f:
    w = csv.DictWriter(f, fieldnames=["titulo", "link", "data"])
    w.writeheader()
    w.writerows(items)
```
